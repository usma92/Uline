```python
# --- USER INPUTS ---
# Path to your Uline spend export (Excel)
INPUT_XLSX = r"C:\Users\dbalan\Desktop\Uline\Uline Spend Detail 2025.xlsx"

# Optional: choose a specific sheet name/index if your workbook has multiple sheets
SHEET_NAME = 0  # e.g., 0 or "Sheet1"

# Optional: where to save charts/csv exports when running locally
OUTPUT_DIR = r"C:\Users\dbalan\Desktop\Uline\output"  # set to None to disable exports

# If True, and if a PO number contains the string "JANET" (any case),
# we will force Ordered By = "JANET" before mapping to a warehouse.
FORCE_JANET_FROM_PO = True
```

```python
import os
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option("display.max_columns", 50)
pd.set_option("display.width", 160)
```

```python
# --- HELPERS (loader + cleaning) ---
def _find_header_row(raw: pd.DataFrame, header_key: str = "Date") -> int:
    """Return the first row index that contains the expected header key.

    Uline exports sometimes include a few lines of report metadata above the real header.
    We detect the header by searching for a cell that exactly equals `header_key`.
    """
    matches = raw.apply(lambda row: row.astype(str).str.strip().eq(header_key).any(), axis=1)
    idx = matches[matches].index
    if len(idx) == 0:
        raise ValueError(
            f"Could not find header row containing '{header_key}'. "
            "Open the spreadsheet and confirm the column headers include 'Date'."
        )
    return int(idx[0])


def load_uline_export(xlsx_path: str, sheet_name=0) -> pd.DataFrame:
    """Load a Uline spend export and return a cleaned DataFrame.

    - Auto-detects the header row (looks for 'Date')
    - Coerces key numeric columns to numbers
    - Parses Date to datetime
    """
    # Read a small preview (no header) to detect the real header row
    preview = pd.read_excel(xlsx_path, sheet_name=sheet_name, header=None)
    header_row = _find_header_row(preview, header_key="Date")

    df = pd.read_excel(xlsx_path, sheet_name=sheet_name, header=header_row).copy()

    # Standardize column names (strip whitespace)
    df.columns = [str(c).strip() for c in df.columns]

    # Parse dates
    if "Date" in df.columns:
        df["Date"] = pd.to_datetime(df["Date"], errors="coerce")

    # Coerce money/qty fields (handle $ and commas if present)
    money_cols = [c for c in ["Ext. Price", "Unit Price"] if c in df.columns]
    for c in money_cols:
        df[c] = (
            df[c]
            .astype(str)
            .str.replace("$", "", regex=False)
            .str.replace(",", "", regex=False)
        )
        df[c] = pd.to_numeric(df[c], errors="coerce").fillna(0)

    if "Qty" in df.columns:
        df["Qty"] = pd.to_numeric(df["Qty"], errors="coerce").fillna(0)

    return df


def normalize_requestor(s: pd.Series) -> pd.Series:
    """Normalize requestor names for mapping consistency."""
    return (
        s.astype(str)
         .str.strip()
         .str.upper()
         .str.replace("’", "'", regex=False)
         .str.replace("'", "", regex=False)
         .str.lstrip("#")
    )


def ensure_output_dir(path: str | None) -> Path | None:
    """Create OUTPUT_DIR if provided; return Path or None."""
    if not path:
        return None
    out = Path(path)
    out.mkdir(parents=True, exist_ok=True)
    return out
```

```python
# --- LOAD DATA ---
df = load_uline_export(INPUT_XLSX, sheet_name=SHEET_NAME)

# Basic derived fields
df["Month"] = df["Date"].dt.to_period("M").dt.to_timestamp()

# Quick sanity checks
display(df.head(5))
print(f"Rows: {len(df):,}")
print(f"Date range: {df['Date'].min()} → {df['Date'].max()}")
```

```python
# --- DATA QUALITY QUICK CHECK ---
# Adjust these column names if your export uses slightly different labels.
required_cols = ["Date", "Ext. Price", "Order #"]
missing = [c for c in required_cols if c not in df.columns]
if missing:
    raise KeyError(f"Missing expected columns: {missing}. Available columns: {list(df.columns)}")

summary = pd.DataFrame(
    {
        "rows": [len(df)],
        "orders": [df["Order #"].nunique()],
        "total_spend": [df["Ext. Price"].sum()],
        "avg_line_spend": [df["Ext. Price"].mean()],
        "pct_missing_date": [df["Date"].isna().mean()],
    }
)

display(summary)
```

## Optional: Map `Ordered By` to your 8 warehouse sites

The export does not explicitly include a warehouse/site column.  
If each person primarily orders for one site, you can roll up spend cleanly by adding a mapping below.

- If you **leave the mapping empty**, the notebook will analyze by `Ordered By`.
- If you **fill in the mapping**, you’ll get charts by `Warehouse` instead.



```python
# --- OPTIONAL MAPPING: Ordered By -> Warehouse ---
# Normalize requestor names for consistent mapping
if "Ordered By" in df.columns:
    df["Ordered By"] = normalize_requestor(df["Ordered By"])
else:
    raise KeyError("Expected a column named 'Ordered By' in the export.")

# Optional rule requested: if PO # contains 'JANET', force Ordered By = 'JANET'
# (This helps when the requestor field is inconsistent but PO carries the real signal.)
if FORCE_JANET_FROM_PO and "PO #" in df.columns:
    janet_mask = df["PO #"].astype(str).str.contains("JANET", case=False, na=False)
    df.loc[janet_mask, "Ordered By"] = "JANET"

# Update this mapping to match your site list / aliases
ordered_by_to_warehouse = {
    "JANET": "Charlotte",
    "SUSAN": "Ontario",
    "JOSE": "Orlando",
    "DALE": "Flowood",
    "DELL": "Flowood",
    "GERALD": "Phoenix",
    "GILBERT": "Oklahoma City",
    "TRAVIS": "Atlanta",
    "TRAVISG": "Atlanta",
    "RICARDO": "El Paso",
    "KENNETH": "Atlanta",
    "JONATHAN": "Atlanta",
}

df["Warehouse"] = df["Ordered By"].map(ordered_by_to_warehouse).fillna(df["Ordered By"])

# Audit unmapped values (helpful when building/maintaining the mapping)
unmapped = sorted(set(df["Ordered By"].unique()) - set(ordered_by_to_warehouse.keys()))
print(f"Unmapped Ordered By values ({len(unmapped)}):")
print(unmapped[:50])

display(df.head(10))
```

## Executive KPIs

```python
kpis = {
    "Total spend": df["Ext. Price"].sum(),
    "Total lines": len(df),
    "Total quantity": df["Qty"].sum(),
    "Unique orders": df["Order #"].nunique(),
    "Unique invoices": df["Invoice"].nunique(),
    "Avg $ per order": df.groupby("Order #")["Ext. Price"].sum().mean(),
    "Median $ per order": df.groupby("Order #")["Ext. Price"].sum().median(),
    "Avg line $": df["Ext. Price"].mean(),
}

pd.Series(kpis).to_frame("Value")
```

## Spend over time (monthly trend)

```python
monthly = df.groupby("Month", as_index=False)["Ext. Price"].sum().sort_values("Month")

plt.figure(figsize=(10,4))
plt.plot(monthly["Month"], monthly["Ext. Price"], marker="o")
plt.title("Total Uline Spend by Month (2025)")
plt.xlabel("Month")
plt.ylabel("Spend ($)")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

monthly
```

## Spend by warehouse / requestor (Pareto)

```python
by_wh = (
    df.groupby("Warehouse", as_index=False)["Ext. Price"]
      .sum()
      .sort_values("Ext. Price", ascending=False)
)
by_wh["CumSpend"] = by_wh["Ext. Price"].cumsum()
by_wh["CumPct"] = by_wh["CumSpend"] / by_wh["Ext. Price"].sum()

plt.figure(figsize=(10,5))
plt.bar(by_wh["Warehouse"], by_wh["Ext. Price"])
plt.xticks(rotation=45, ha="right")
plt.title("Spend by Warehouse")
plt.ylabel("Spend ($)")
plt.tight_layout()
plt.show()

plt.figure(figsize=(10,4))
plt.plot(by_wh["Warehouse"], by_wh["CumPct"], marker="o")
plt.xticks(rotation=45, ha="right")
plt.title("Cumulative % of Spend (Pareto)")
plt.ylabel("Cumulative %")
plt.ylim(0, 1.05)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

by_wh.head(30)
```

## Heatmap: spend by month × warehouse/requestor

```python
pivot = df.pivot_table(
    index="Warehouse",
    columns="Month",
    values="Ext. Price",
    aggfunc="sum",
    fill_value=0
)

plt.figure(figsize=(12,6))
plt.imshow(pivot.values, aspect="auto")
plt.yticks(range(len(pivot.index)), pivot.index)
plt.xticks(range(len(pivot.columns)), [d.strftime("%b") for d in pivot.columns], rotation=0)
plt.title("Spend Heatmap (rows: warehouse/requestor, cols: month)")
plt.xlabel("Month (2025)")
plt.ylabel("Warehouse")
plt.colorbar(label="Spend ($)")
plt.tight_layout()
plt.show()

pivot
```

## Category analysis (top categories + monthly mix)

```python
top_cat = (
    df.groupby("Category", as_index=False)["Ext. Price"]
      .sum()
      .sort_values("Ext. Price", ascending=False)
)

plt.figure(figsize=(10,5))
plt.bar(top_cat["Category"].head(15), top_cat["Ext. Price"].head(15))
plt.xticks(rotation=45, ha="right")
plt.title("Top 15 Categories by Spend")
plt.ylabel("Spend ($)")
plt.tight_layout()
plt.show()

# Monthly category mix (top 8 categories + 'Other')
top8 = top_cat["Category"].head(8).tolist()
df["Category_top"] = np.where(df["Category"].isin(top8), df["Category"], "Other")
mix = df.pivot_table(index="Month", columns="Category_top", values="Ext. Price", aggfunc="sum", fill_value=0).sort_index()

plt.figure(figsize=(12,5))
# Stacked area
plt.stackplot(mix.index, mix.T.values, labels=mix.columns)
plt.title("Monthly Spend Mix (Top Categories + Other)")
plt.xlabel("Month")
plt.ylabel("Spend ($)")
plt.legend(loc="upper left", bbox_to_anchor=(1.01, 1.0))
plt.tight_layout()
plt.show()

top_cat.head(25)
```

## Item / SKU analysis (Model #)

## Savings opportunity flags (rule-based)

These are **starting points**—they highlight where to investigate:
- **Free offers** / zero-dollar lines: validate if they’re useful or avoid clutter
- **High-frequency SKUs**: candidate for standardization, min/max, bulk buying, or vendor-managed inventory
- **“Long tail” suppliers/categories**: consolidate to fewer SKUs or brands
- **Outlier orders**: ensure they were planned/approved (one-time equipment vs. drift)


```python
# Free / $0 lines
free_lines = df[df["Ext. Price"].fillna(0) <= 0].copy()

# High-frequency SKUs
hi_freq = (
    df.groupby(["Model #", "Description"])
      .agg(lines=("Ext. Price", "size"),
           orders=("Order #", "nunique"),
           total_spend=("Ext. Price", "sum"),
           avg_line_spend=("Ext. Price", "mean"))
      .reset_index()
      .sort_values(["lines", "total_spend"], ascending=False)
)

# Unique models per category
models_per_cat = (
    df.groupby('Category')['Model #'].nunique()
        .reset_index().rename(columns={'Model #': 'unique_models'})
        .sort_values('unique_models', ascending=False)
)


# Outlier orders (top 3%)
threshold = order_totals["Ext. Price"].quantile(0.97)
outliers = order_totals[order_totals["Ext. Price"] >= threshold]

print(f"Free/$0 lines: {len(free_lines)}")
display(free_lines.head(20))

print(f"Hi freq lines: {len(hi_freq)}")
print("Top 5 high-frequency SKUs:")
for _, row in hi_freq.head(5).iterrows():
    print(f"- {row['Model #']}: {row['Description']} ({row['lines']} lines)")
display(hi_freq.head(25))

print(f"Top 3% order threshold: ${threshold:,.2f} (orders >= threshold: {len(outliers)})")
print(f"Outliers: {len(outliers)}")
display(outliers.head(25))

print(f"Unique models per category: {len(models_per_cat)}")
display(models_per_cat.head(25))
```

```python
# Top 5 corrugated box models by warehouse
corrugated = df[df['Category'].str.contains('Corrugated|Boxes', case=False, na=False)]
top_corrugated = corrugated.groupby(['Warehouse', 'Model #', 'Description'])['Ext. Price'].sum().reset_index().sort_values(['Warehouse', 'Ext. Price'], ascending=[True, False])
top_5_per_wh = top_corrugated.groupby('Warehouse').head(5)
top_5_per_wh

import matplotlib.pyplot as plt
for warehouse in top_5_per_wh['Warehouse'].unique():
    subset = top_5_per_wh[top_5_per_wh['Warehouse'] == warehouse]
    plt.figure(figsize=(8,4))
    plt.bar(subset['Model #'], subset['Ext. Price'])
    plt.title(f'Top 5 Corrugated Box Models for {warehouse}')
    plt.xlabel('Model #')
    plt.ylabel('Total Spend ($)')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```
